```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS

nlp = spacy.load("en_core_web_sm")

def clean_and_extract_tags(text):
    doc = nlp(text.lower())
    tags = [token.text for token in doc if token.text.isalnum() and token.text not in STOP_WORDS]
    return ', '.join(tags)

columns_to_extract_tags_from = ['prod_name', 'product_type_name',
       'product_group_name', 'perceived_colour_master_name', 'department_name',
       'index_group_name', 'section_name',]

for column in columns_to_extract_tags_from:
    hm_data[column] = hm_data[column].apply(clean_and_extract_tags)
```

```python
# Concatenate the cleaned tags from all relevant columns
hm_data['Tags'] = hm_data[columns_to_extract_tags_from].apply(lambda row: ', '.join(row), axis=1)
```

```python
def search_similar_items(user_input, merged_dataset, tfidf_vectorizer, tfidf_matrix_content, top_n=10):
    # Vectorize the user input
    user_input_tfidf = tfidf_vectorizer.transform([user_input])

    # Compute cosine similarity between user input and the existing items
    cosine_similarities = cosine_similarity(user_input_tfidf, tfidf_matrix_content).flatten()

    # Get the indices of the top N most similar items
    similar_item_indices = cosine_similarities.argsort()[-top_n:][::-1]

    # Extract the most similar items' details
    similar_items = merged_dataset.iloc[similar_item_indices][['article_id', 'prod_name', 'product_type_name',
       'product_group_name', 'perceived_colour_master_name', 'department_name',
       'index_group_name', 'section_name', 'rating', 't_dat', 'customer_id']]
    
    
    
    # Print the results in table format
    print(similar_items.to_string(index=False))  # Display the table without index numbers

    return similar_items

# Example usage:
user_input = "Shirt black men"  # User provides tags
similar_items = search_similar_items(user_input, hm_data, tfidf_vectorizer, tfidf_matrix_content, top_n=5)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from implicit.als import AlternatingLeastSquares
import scipy.sparse as sparse

# 1. Load Data (merged_dataset should be your combined influencer data)
# Sample merged data (replace this with your actual merged_dataset)

# Prepare Interaction Matrix (for ALS input)
interaction_data = hm_data.pivot_table(index='customer_id', columns='prod_name', values='rating', aggfunc='sum').fillna(0)

# Convert the interaction matrix to a sparse matrix (required for ALS)
interaction_sparse = sparse.csr_matrix(interaction_data.values)

# 2. ALS Model
# Initialize the ALS model (using implicit feedback)
als_model = AlternatingLeastSquares(factors=50, regularization=0.01, iterations=30)

# Train the ALS model
als_model.fit(interaction_sparse)

# 3. Recommend Similar Influencers based on ALS
def recommend_similar_influencers_als(influencer_name, interaction_data, model, top_n=5):
    if influencer_name not in interaction_data.index:
        print(f"{influencer_name} not found in interaction data.")
        return pd.DataFrame()

    # Get the index of the influencer in your interaction data
    influencer_idx = interaction_data.index.get_loc(influencer_name)

    # Use ALS to get similar items based on the influencer index
    influencer_recommendations = model.similar_items(influencer_idx, N=top_n + 1)  # ALS returns (index, score) pairs

    # Extract recommended influencer indices and similarity scores
    recommendations = influencer_recommendations  # Exclude the first result (self-recommendation)

    # Map ALS indices back to influencer names using interaction_data
    recommended_influencers = [interaction_data.index[int(i[0])] for i in recommendations if int(i[0]) < len(interaction_data.index)]
    scores = [i[1] for i in recommendations]  # i[1] is the score

    # Return a DataFrame with the recommended influencers and their similarity scores
    return pd.DataFrame({
        'Influencer_name': recommended_influencers,
        'ALS_Similarity': scores
    })




# Example usage for ALS-based collaborative filtering
als_collab_influencers = recommend_similar_influencers_als('denim', interaction_data, als_model, top_n=5) #product name
print("\nALS-Based Collaborative Filtering Recommendations:")
print(als_collab_influencers)

# 4. Content-Based Similarity (same as your original code)
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix_content = tfidf_vectorizer.fit_transform(hm_data['Tags'])
cosine_similarities_content = cosine_similarity(tfidf_matrix_content, tfidf_matrix_content)

def search_similar_items_content(user_input, merged_dataset, tfidf_vectorizer, tfidf_matrix_content, top_n=5):
    user_input_tfidf = tfidf_vectorizer.transform([user_input])
    cosine_similarities = cosine_similarity(user_input_tfidf, tfidf_matrix_content).flatten()
    similar_item_indices = cosine_similarities.argsort()[-top_n:][::-1]
    return merged_dataset.iloc[similar_item_indices]

# Example user input for content-based search
user_input_content = "Denim Men "
similar_items_content = search_similar_items_content(user_input_content, hm_data, tfidf_vectorizer, tfidf_matrix_content, top_n=6)
print("\nContent-Based Recommendations:")
print(similar_items_content[['prod_name', 'product_type_name',
       'product_group_name', 'perceived_colour_master_name', 'department_name',
       'index_group_name', 'section_name', 'rating', 'Tags']])

# 5. Merge Recommendations
merged_recommendations = pd.concat([similar_items_content[['article_id', 'prod_name', 'product_type_name',
       'product_group_name', 'perceived_colour_master_name', 'department_name',
       'index_group_name', 'section_name', 'rating', 't_dat', 'customer_id']],
                                     als_collab_influencers], ignore_index=True)

print("\nMerged Recommendations:")
print(merged_recommendations)

# Save to CSV
merged_recommendations.to_csv('als_hnm_result.csv')
```

```python
import numpy as np
from scipy.sparse import csr_matrix
from scipy.sparse.linalg import svds

import numpy as np
from scipy.sparse import csr_matrix
from scipy.sparse.linalg import svds

def svd_collaborative_filtering(interaction_data, top_n=5):
    # Fill missing values with 0
    interaction_data = interaction_data.fillna(0)

    # Center the matrix by subtracting the mean
    interaction_data_mean = interaction_data.mean(axis=1).values.reshape(-1, 1)
    normalized_data = interaction_data - interaction_data_mean

    # Convert to sparse matrix if large or sparse
    interaction_data_sparse = csr_matrix(normalized_data.values)

    # Set k to be less than the smallest dimension of the matrix
    num_users, num_items = interaction_data_sparse.shape
    k = min(num_users, num_items) - 1  # Make sure k is smaller than the smallest dimension

    # Perform SVD
    U, sigma, Vt = svds(interaction_data_sparse, k=k)

    # Convert sigma to a diagonal matrix
    sigma = np.diag(sigma)

    # Make predictions
    predicted_ratings = np.dot(np.dot(U, sigma), Vt) + interaction_data_mean

    # Create a DataFrame for predicted ratings
    predicted_ratings_df = pd.DataFrame(predicted_ratings, index=interaction_data.index, columns=interaction_data.columns)

    return predicted_ratings_df


# Example usage for SVD-based collaborative filtering
svd_predictions = svd_collaborative_filtering(interaction_data)
print("\nSVD Collaborative Filtering Predictions:")
print(svd_predictions)

# 3. Content-Based Similarity
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix_content = tfidf_vectorizer.fit_transform(hm_data['Tags'])
cosine_similarities_content = cosine_similarity(tfidf_matrix_content, tfidf_matrix_content)

def search_similar_items_content(user_input, hm_data, tfidf_vectorizer, tfidf_matrix_content, top_n=5):
    user_input_tfidf = tfidf_vectorizer.transform([user_input])
    cosine_similarities = cosine_similarity(user_input_tfidf, tfidf_matrix_content).flatten()
    similar_item_indices = cosine_similarities.argsort()[-top_n:][::-1]
    return hm_data.iloc[similar_item_indices]

# Example user input for content-based search
user_input_content = "Denim, Men"
similar_items_content = search_similar_items_content(user_input_content, hm_data, tfidf_vectorizer, tfidf_matrix_content, top_n=6)

# 4. Recommend Similar Influencers based on SVD collaborative filtering
def recommend_similar_influencers_svd(influencer_name, svd_predictions, top_n=5):
    if influencer_name not in svd_predictions.index:
        return pd.DataFrame()

    influencer_recommendations = svd_predictions.loc[influencer_name].sort_values(ascending=False)
    return influencer_recommendations.head(top_n)

# Example usage for SVD-based collaborative filtering
svd_collab_influencers = recommend_similar_influencers_svd('Denim', svd_predictions, top_n=5)
print("\nSVD-Based Collaborative Filtering Recommendations:") #enter product name
print(svd_collab_influencers)

# 5. Merge Recommendations
merged_recommendations = pd.concat([similar_items_content[['article_id', 'prod_name', 'product_type_name',
       'product_group_name', 'perceived_colour_master_name', 'department_name',
       'index_group_name', 'section_name', 'rating', 't_dat', 'customer_id']],
                                     svd_collab_influencers.reset_index().rename(columns={'article_id': 'prod_name', 0: 'SVD_Similarity'})],
                                    ignore_index=True)

print("\nMerged Recommendations:")
print(merged_recommendations)

# Save the merged recommendations to a CSV file
merged_recommendations.to_csv('svd_hnm_result.csv', index=False)


```
